# WIN BY RUNS

```python
batting_first = ipl[ipl['win_by_runs']!=0]
batting_first.head()
```

# RUNS MARGINE WHILE BATTING FIRST

```python
plt.hist(batting_first['win_by_runs'])
plt.xlabel('runs')
plt.ylabel('matches')
plt.title('Run Margins')
```

# MOST WINS WHILE BATTING FIRST

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(20,10))
plt.bar(list(batting_first['winner'].value_counts().keys()[0:5]) , list(batting_first['winner'].value_counts()[0:5]) , color = ['yellow','blue', 'purple' , 'orange' , 'red' ] )
plt.title('MOST WINS WHILE BATTING FIRST')
```

```python
plt.figure(figsize=(20,10))
plt.pie(batting_first['winner'].value_counts() , labels = batting_first['winner'].value_counts().keys() , autopct='%0.1f%%')
```

# STATS WHILE BATTING SECOND

```python
secondBatting= ipl[ipl['win_by_wickets']!=0]
secondBatting.head()
```

```python
plt.figure(figsize=(20,10))
plt.hist(secondBatting['win_by_wickets'],bins=30)
plt.title('WICKETS MARGINS')
```

# MOST WINS WHILE BATTING SECOND

```python
plt.figure(figsize=(20,10))
a1 = secondBatting['winner'].value_counts()[0:5]
a2 = secondBatting['winner'].value_counts()[0:5].keys()
plt.bar(a2,a1 , color = ('r' ,'purple' , 'blue' , 'blue' , 'blue'))
plt.title('Most wins by wickets')
```

```python
plt.figure(figsize=(20,10))
plt.pie(secondBatting['winner'].value_counts(),labels = secondBatting['winner'].value_counts().keys() , autopct = '%0.1f%%')
```

```python
plt.figure(figsize=(20,10))
a3 = secondBatting['player_of_match'].value_counts()[0:5]
a4 = secondBatting['player_of_match'].value_counts()[0:5].keys()
plt.bar(a4,a3 , color = 'orange')
plt.title('Most MoMs')
```
